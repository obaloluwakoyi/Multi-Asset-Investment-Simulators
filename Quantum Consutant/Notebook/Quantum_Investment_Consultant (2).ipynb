```python
# --- Step 1: Client Intake ---
print("--- Quantum Investment Consultant ---")
user_input = input("Enter stock tickers (e.g., AAPL, TSLA, MSFT, BTC-USD): ")
tickers = [t.strip().upper() for t in user_input.split(",")]

# Pulling 1 year of historical data using yfinance
import yfinance as yf
start_date = datetime(2023, 1, 1)
end_date = datetime(2024, 1, 1)

data_yf = yf.download(tickers, start=start_date, end=end_date)['Close']

# Calculate daily returns
returns = data_yf.pct_change().dropna()

# Calculate expected returns (mu) and covariance matrix (sigma)
# Qiskit's PortfolioOptimization expects mu as a 1D array and sigma as a 2D array.
mu = returns.mean().values
sigma = returns.cov().values

# --- Step 2: Set Risk & Diversification ---
client_risk = float(input("Risk Aversion (1=Aggressive, 10=Conservative): "))
q_risk_factor = max(1, min(10, client_risk)) / 10.0

q_budget = int(input(f"Max number of assets to pick (1-{len(tickers)}): "))

# Setup Problem
portfolio = PortfolioOptimization(
    expected_returns=mu, covariances=sigma,
    risk_factor=q_risk_factor, budget=q_budget
)
qp = portfolio.to_quadratic_program()

# --- Step 3: Quantum Execution ---
vqe = SamplingVQE(sampler=Sampler(), ansatz=EfficientSU2(qp.get_num_vars(), reps=1), optimizer=COBYLA())
optimizer = MinimumEigenOptimizer(vqe)
result = optimizer.solve(qp)

# --- Step 4: Final Allocation & Export ---
selection = portfolio.interpret(result)
selected_tickers = [tickers[i] for i, val in enumerate(selection) if val == 1]

print(f"\n✅ Recommended Assets: {selected_tickers}")

investment = float(input("Total Investment Amount ($): "))
amt_per_stock = investment / len(selected_tickers) if selected_tickers else 0

# Create Report
report_df = pd.DataFrame({
    "Ticker": selected_tickers,
    "Allocation_USD": [amt_per_stock] * len(selected_tickers),
    "Recommendation_Date": datetime.now().strftime("%Y-%m-%d")
})

# Save to CSV
report_df.to_csv("Portfolio_Recommendation.csv", index=False)
print("\n📁 Recommendation saved to 'Portfolio_Recommendation.csv'")
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import plotly.express as px
from datetime import datetime, timedelta
from qiskit_finance.applications.optimization import PortfolioOptimization
from qiskit_algorithms import SamplingVQE
from qiskit_algorithms.optimizers import COBYLA
from qiskit.circuit.library import EfficientSU2
from qiskit.primitives import StatevectorSampler as Sampler
from qiskit_optimization.algorithms import MinimumEigenOptimizer
from qiskit_finance.data_providers import RandomDataProvider
```

```python
# ==========================================
# 1. DATA ACQUISITION
# ==========================================
print("--- 🚀 Quantum Alpha Consultant v3.0 ---")
user_input = input("Enter Tickers (e.g., AAPL, TSLA, MSFT, AMD, NVDA): ")
tickers = [t.strip().upper() for t in user_input.split(",") if t.strip()]

end_date = datetime.now()
start_date = end_date - timedelta(days=365)

data = yf.download(tickers, start=start_date, end=end_date)['Close']
returns = data.pct_change().dropna()
mu, sigma = returns.mean().values, returns.cov().values

# ==========================================
# 2. QUANTUM OPTIMIZATION
# ==========================================
risk_val = float(input("Risk Aversion (1-10): "))
q_budget = int(input("Allocation Units (10-20): "))
investment_total = float(input("Total Investment ($): "))

qp = PortfolioOptimization(expected_returns=mu, covariances=sigma,
                           risk_factor=risk_val/10, budget=q_budget).to_quadratic_program()

ansatz = EfficientSU2(qp.get_num_vars(), reps=2, entanglement="full")
vqe = SamplingVQE(sampler=Sampler(), ansatz=ansatz, optimizer=COBYLA(maxiter=200))
selection = PortfolioOptimization(expected_returns=mu, covariances=sigma,
                                  risk_factor=risk_val/10, budget=q_budget).interpret(MinimumEigenOptimizer(vqe).solve(qp))

# ==========================================
# 3. METRICS & SIMULATION
# ==========================================
df_report = pd.DataFrame([{"Ticker": tickers[i], "Weight": units/q_budget, "Amount": investment_total*(units/q_budget)}
                         for i, units in enumerate(selection) if units > 0])

active_rets = (returns[df_report['Ticker']] * df_report['Weight'].values).sum(axis=1)
port_cum = (1 + active_rets).cumprod()

# --- Monte Carlo Simulation (1 Year Outlook) ---
t_intervals = 252
iterations = 1000
daily_avg = active_rets.mean()
daily_std = active_rets.std()

# Geometric Brownian Motion
simulations = np.zeros((t_intervals, iterations))
for i in range(iterations):
    sim_rets = np.random.normal(daily_avg, daily_std, t_intervals)
    simulations[:, i] = investment_total * (1 + sim_rets).cumprod()

# ==========================================
# 4. INTERACTIVE VISUALS
# ==========================================
# Performance & Correlation
fig_perf = go.Figure([go.Scatter(x=port_cum.index, y=port_cum, name='Quantum Portfolio', line=dict(color='#00FFAA'))])
fig_heat = px.imshow(returns[df_report['Ticker']].corr(), text_auto=".2f", color_continuous_scale='Viridis', title="Asset Correlation")

# Monte Carlo Plot
fig_mc = go.Figure()
for i in range(min(iterations, 100)): # Plot 100 sample paths
    fig_mc.add_trace(go.Scatter(y=simulations[:, i], mode='lines', line=dict(width=0.5), opacity=0.3, showlegend=False))
fig_mc.update_layout(title=f"Monte Carlo: 1,000 Future Paths (${investment_total} Base)", xaxis_title="Trading Days", yaxis_title="Portfolio Value")

# Convert to HTML
perf_h = fig_perf.to_html(full_html=False, include_plotlyjs='cdn')
heat_h = fig_heat.to_html(full_html=False, include_plotlyjs=False)
mc_h = fig_mc.to_html(full_html=False, include_plotlyjs=False)
pie_h = go.Figure(data=[go.Pie(labels=df_report['Ticker'], values=df_report['Weight'], hole=0.4)]).to_html(full_html=False, include_plotlyjs=False)

# ==========================================
# 5. FINAL DASHBOARD
# ==========================================
html_template = f"""
<!DOCTYPE html>
<html>
<head>
    <style>
        body {{ font-family: sans-serif; background: #0f172a; color: #f8fafc; padding: 40px; }}
        .card {{ background: #1e293b; padding: 20px; border-radius: 15px; margin-bottom: 20px; border: 1px solid #334155; }}
        .grid {{ display: grid; grid-template-columns: 1fr 1fr; gap: 20px; }}
        .metric {{ font-size: 32px; color: #00FFAA; font-weight: bold; }}
        table {{ width: 100%; border-collapse: collapse; margin-top: 10px; }}
        th {{ text-align: left; color: #94a3b8; }}
        td {{ padding: 10px 0; border-bottom: 1px solid #334155; }}
    </style>
</head>
<body>
    <h1>Quantum Portfolio Intelligence Dashboard</h1>
    <div class="grid">
        <div class="card">
            <h2>Live Metrics</h2>
            <div class="metric">{(port_cum.iloc[-1]-1)*100:.2f}%</div>
            <p>Historical Annual Return</p>
            <div class="metric">${simulations[-1, :].mean():,.2f}</div>
            <p>Expected 1-Year Value (MC Mean)</p>
        </div>
        <div class="card">
            <h2>Allocation</h2>
            {df_report[['Ticker', 'Weight', 'Amount']].to_html(index=False)}
        </div>
        <div class="card">{perf_h}</div>
        <div class="card">{pie_h}</div>
        <div class="card">{heat_h}</div>
        <div class="card">{mc_h}</div>
    </div>
</body>
</html>
"""

with open("Quantum_Alpha_Dashboard.html", "w") as f: f.write(html_template)
print("\n✅ Terminal Output: Strategy Optimized. View 'Quantum_Alpha_Dashboard.html' for the full report.")
```
